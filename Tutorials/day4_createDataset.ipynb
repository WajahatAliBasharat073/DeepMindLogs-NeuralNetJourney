```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Generate 100 rows of data
num_rows = 100

# Generate CGPA (0 to 10, float)
cgpa = np.round(np.random.uniform(0, 10, num_rows), 2)

# Generate Resume Score (0 to 10, float)
resume_score = np.round(np.random.uniform(0, 10, num_rows), 2)

# Generate Placement (0 or 1, balanced)
placement = np.random.choice([0, 1], size=num_rows, p=[0.5, 0.5])

# Create a DataFrame
data = {
    'cgpa': cgpa,
    'resume_score': resume_score,
    'placement': placement
}

df = pd.DataFrame(data)

# Shuffle the dataset to ensure randomness
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Display the first few rows of the dataset
print(df.head())

# Save the dataset to a CSV file (optional)
df.to_csv('placement_dataset.csv', index=False)
```
